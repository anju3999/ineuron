2.Implement a Python program to generate all sentences where subject is in ["Americans", "Indians"] and verb is in ["Play", "watch"] and the object is in ["Baseball","cricket"].


```python
subject=["Americans","Indians"]
verb=["play","watch"]
object=["Baseball","Cricket"]

for s in subject:
    for v in verb:
        for o in object:
            print("{} {} {}.".format(s,v,o))
```

# Task 2:
1. Write a function so that the columns of the output matrix are powers of the input vector.


```python
def generate_matrix(l, r=5, descending=False):
    if descending==True:
        return [[i**j for j in range(r, -1, -1)] for i in l]
    return [[i**j for j in range(r+1)] for i in l]
```

```python
generate_matrix([1,2,3,4,5,6])
```

```python
generate_matrix([1,2,3,4,5,6],descending=True)
```
